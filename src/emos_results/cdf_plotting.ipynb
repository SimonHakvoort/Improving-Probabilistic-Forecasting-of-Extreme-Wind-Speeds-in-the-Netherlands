```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import tensorflow as tf
```

```python
a = 5
b = -.6
x = np.linspace(0, 20, 1000)
y = 1 / (1 + np.exp(-(a + b * x)))
plt.plot(x, y)
plt.xlim(0,20)
plt.ylim(0,1)
plt.grid()
```

```python
# plot a normal cdf with mean 13 and std 4
x = np.linspace(0, 25, 100)

mean = 13
std = 2
c = 0.03

y = norm.cdf(x, mean, std) + c

plt.plot(x, y)
plt.xlim(0, 25)
plt.ylim(0, 1 + c)
plt.grid()
plt.show()
```
